```python
import  pandas as pd
import requests
import json
import re
import numpy as np
```

```python
map_data=pd.read_csv("map_data6.csv")
```

```python
map_lngs=[]
map_lats=[]
map_district=[]
```

```python
map_data["map_infos"]=map_data["map_infos"].fillna("error")
map_data["map_infos"]=map_data["map_infos"].str.replace('\2','/2')
map_data["map_infos"]=map_data["map_infos"].str.replace('\\','/')
map_data["map_infos"]=map_data["map_infos"].str.replace('""贵阳市花溪区""','"贵阳市花溪区"')
map_data["map_infos"]=map_data["map_infos"].str.replace('""贵阳市小河区""','"贵阳市小河区"')
```

```python
# print(map_data.loc[4676,"map_infos"])
```

```python
for index,map_info in enumerate(map_data["map_infos"]):
    print(index)
    # map_info.replace('""','"')
    if(map_info=="error"):
        map_lats.append(26.33)
        map_lngs.append(106.62)
        map_district.append("无")
    else:
        district=re.findall(r"\w\w区",map_info)
        if(len(district)!=0):
            map_district.append(district[0])
        else:
            map_district.append("无")
        try:
            map_info = json.loads(map_info)
        except(json.JSONDecodeError) :
            print("错误")
            map_info=map_info.replace('""','"')
            # map_info=map_info.replace('"""','"')
            # map_info=map_info.replace("\\",'/')
            # map_info=map_info.replace("\\2",'/2') 
            map_info = json.loads(map_info)
        
        if (len(map_info['results']) != 0):
            map_lat = map_info['results'][0]["location"]['lat']
            map_lng = map_info['results'][0]["location"]['lng']
            print(map_lat)
            print(map_lng)
            print('\n')
            map_lats.append(map_lat)
            map_lngs.append(map_lng)
        else:
            print("第" + str(index) + "个错误")
            map_lats.append(26.33)
            map_lngs.append(106.62)
```

```python
print(len(map_lngs))
print(len(map_district))
```

```python
map_info=pd.DataFrame()
map_info["ID"]=[i for i in range(50000,56651)]
map_info["lngs"]=map_lngs
map_info["lat"]=map_lats
# print(len(map_lngs))
# print(len(map_district))
map_info["district"]=map_district
print(map_info.tail())
map_infos=pd.read_csv("map_info.csv",encoding="utf-8")
# map_infos=map_infos.drop("ID.1",axis=1)
print(map_infos.tail())
map_infos=map_infos.append(map_info,ignore_index=True)
print(map_infos.tail())
```

```python
map_infos.to_csv("map_info.csv",encoding="utf-8",index=False)
```
